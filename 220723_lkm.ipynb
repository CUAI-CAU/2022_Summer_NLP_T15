# 2. 데이터를 하나로 합치기 위해 간단한 전처리 & 데이터 병합

```python
aihub_train1 = aihub_train.copy()
aihub_train2 = aihub_train.copy()
naver_train1 = naver_train.copy()

aihub_train1 = aihub_train1[['번호', '감정_대분류', '사람문장1']]
aihub_train2 = aihub_train2[['번호', '감정_대분류', '사람문장2']]
```

```python
aihub_train1.head(3)
```

```python
aihub_train2.head(3)
```

aihub_train1과 aihub_train2의 아래쪽 Null 데이터들을 미리 한꺼번에 제거

```python
print(aihub_train1.isnull().sum(), aihub_train2.isnull().sum())
```

```python
aihub_train1.dropna(axis=0, inplace=True)
aihub_train2.dropna(axis=0, inplace=True)

print(aihub_train1.isnull().sum(), aihub_train2.isnull().sum())
```

```python
# 확인
aihub_train1.tail(3)
```

aihub의 열 이름을 naver 영화평점의 열 이름으로 통일!

```python
aihub_train1.columns = ['id', 'label', 'document']
aihub_train1.head(3)
```

```python
aihub_train2.columns = ['id', 'label', 'document']
aihub_train2.head(3)
```

aihub의 label 데이터 중 기쁨(긍정)은 1, 나머지 감정들(부정)은 0으로 변경 -> <u>replace()</u> 사용! 

```python
aihub_train1.replace('기쁨', 1, inplace=True)
aihub_train1.replace(['분노','슬픔','불안','상처','당황'], [0,0,0,0,0], inplace=True)

aihub_train2.replace('기쁨', 1, inplace=True)
aihub_train2.replace(['분노','슬픔','불안','상처','당황'], [0,0,0,0,0], inplace=True)
```

```python
aihub_train1.head(3)
```

```python
# aihub_train1, aihub_train2의 id, label 데이터타입을 int로 변경
aihub_train1[['id', 'label']].astype(int)
aihub_train2[['id', 'label']].astype(int)
```

```python
# error를 통해 감정 중 '불안'이 뒤에 공백이 있어서 아직 변환이 덜 된 것을 확인
aihub_train1.replace('불안 ', 0, inplace=True)
aihub_train2.replace('불안 ', 0, inplace=True)
```

```python
aihub_train1[['id', 'label']].astype(int)
aihub_train2[['id', 'label']].astype(int)
```
